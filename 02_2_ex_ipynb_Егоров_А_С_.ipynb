Записать данные из файла в массив

````python
import pandas as pd
grades_csv = pd.read_csv('grades.csv', delimiter = ';')
grades = pd.DataFrame(grades_csv)
grades.columns = ['name', 'grades_bykva']
grades
````

Добавьте средний балл учащегося в строку учащегося (A=4,B=3,C=2,D=1).

````python
grades_mapping = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
grades['grades_numeric'] = grades['grades_bykva'].replace(grades_mapping)
print(grades)
````

Запишите этот файл в новый файл grades_gpa.csv.

````python
import csv
df = pd.DataFrame(grades, columns=['name', 'grades_bykva', 'grades_numeric'])
df.to_csv('grades_gpa.csv', index=False)
````

Импорт библиотеки Pandas, создание датафрейма с именем hit_albums

````python
import pandas as pd
hit_albums_csv = pd.read_csv("hit_albums.csv")
hit_albums = pd.DataFrame(hit_albums_csv)
````

Расчет средних сертифицированных продаж для всех альбомов

````python
average_sales = hit_albums['Certified sales (millions)'].mean()
print(f'Средние сертифицированные продажи для всех альбомов: {round(average_sales, 2)}')
````

* Создайте новый фрейм данных, содержащий только альбомы с сертифицированными продажами (более 20 миллионов человек купили альбом).

````python
new_df = hit_albums[hit_albums['Certified sales (millions)'] > 20]
new_df
````


 * Создайте новый фрейм данных на основе набора данных hit_albums, который содержит только исполнителей, у которых в списке есть как минимум два альбома.

Группировка по исполнителям и подсчет количества альбомов

````python
albums_count = hit_albums.groupby('Artist').size()
albums_count
````

Фильтрация по исполнителям с двумя или более альбомами

````python
multiple_albums_artists = albums_count[albums_count >= 2].reset_index()
print(multiple_albums_artists)
````

* Создайте новый фрейм данных, содержащий общую сумму всех сертифицированных продаж за каждый год.

````python
total_sales_per_year = hit_albums.groupby('Released')['Certified sales (millions)'].sum().reset_index()
print(total_sales_per_year)
````
